```python
import os
from IPython.display import Image

# Step 1: Get latest detection folder
detect_base = '/content/yolov5/runs/detect'
output_dirs = sorted(os.listdir(detect_base))
latest_folder = output_dirs[-1]
result_dir = os.path.join(detect_base, latest_folder)

# Step 2: List files inside the result folder
files = os.listdir(result_dir)
print("Files in detection output:", files)

# Step 3: Display the first image file found
for f in files:
    if f.lower().endswith(('.jpg', '.jpeg', '.png')):
        display_path = os.path.join(result_dir, f)
        display(Image(filename=display_path))
        break
else:
    print("‚ö†Ô∏è No image file found in detection folder.")

```
