We'll use a text classification model like bert-base-uncased.

```python
from transformers import pipeline

tagger = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

 Define Possible Tags

```python
categories = ["Python", "Machine Learning", "Deep Learning", "Web Development", "Data Science", "Cybersecurity"]
```

 Multi-label=True allows multiple relevant tags
 The output will rank tags based on confidence score

```python
question = "How do I train a neural network using PyTorch?"
result = tagger(question, candidate_labels=categories, multi_label=True)

# Print tags with confidence scores
for tag, score in zip(result["labels"], result["scores"]):
    print(f"{tag}: {score:.2f}")
```

```python
questions = [
    "What is the best way to learn Python?",
    "How does backpropagation work in deep learning?",
    "What are the security risks in web applications?"
]

for q in questions:
    result = tagger(q, candidate_labels=categories, multi_label=True)
    top_tags = [result["labels"][i] for i in range(len(result["scores"])) if result["scores"][i] > 0.5]  # Only keep high-confidence tags
    print(f"Question: {q}\nTags: {', '.join(top_tags)}\n")
```

```python
questions = [
    "How do I implement a neural network using TensorFlow?",
    "What are the best practices for securing a web server?",
    "How can I scrape data from a website using Python?",
    "What is the difference between supervised and unsupervised learning?",
    "How do I build a chatbot using natural language processing?"
]

categories = ["Python", "Machine Learning", "Deep Learning", "Web Development", "Data Science", "Cybersecurity", "NLP", "AI"]

for q in questions:
    result = tagger(q, candidate_labels=categories, multi_label=True)
    top_tags = [result["labels"][i] for i in range(len(result["scores"])) if result["scores"][i] > 0.5]  # Keep high-confidence tags
    print(f"📌 **Question:** {q}\n🏷️ **Tags:** {', '.join(top_tags)}\n")
```
