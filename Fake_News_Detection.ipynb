# Define the prediction function

```python
from transformers import pipeline
import gradio as gr


fake_news_detector = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")


def detect_fake_news(text):
    result = fake_news_detector(text)[0]
    label = result['label']  # Labels could be "POSITIVE" or "NEGATIVE"
    score = result['score']  # Confidence score
    return f"Prediction: {label} (Confidence: {score:.2f})"

# Gradio interface
interface = gr.Interface(fn=detect_fake_news, inputs="text", outputs="text", title="Fake News Detector")
interface.launch()
```
