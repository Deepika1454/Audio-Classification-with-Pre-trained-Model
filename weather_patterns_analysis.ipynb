```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = {
    'Date': pd.date_range(start='2020-01-01', periods=365, freq='D'),
    'Temperature': [20 + (x % 30) + (x % 7) for x in range(365)],
    'Rainfall': [0 if x % 10 != 0 else (x % 20) for x in range(365)]
}

weather_df = pd.DataFrame(data)
```

```python

print(weather_df.head())
```

Line Plot for Temperature:

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=weather_df, x='Date', y='Temperature', color='orange')
plt.title('Daily Temperature Over One Year')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

Bar Plot for Rainfall

```python
plt.figure(figsize=(14, 6))
sns.barplot(data=weather_df, x='Date', y='Rainfall', color='blue')
plt.title('Daily Rainfall Over One Year')
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
