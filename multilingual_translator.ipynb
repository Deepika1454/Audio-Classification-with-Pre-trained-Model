```python
from transformers import MarianMTModel, MarianTokenizer

def translate_text(text, model_name):
    """
    Translate text using a pre-trained model.
    Args:
      text (str): Text in English to translate.
      model_name (str): Model name for the desired language pair.

    Returns:
      str: Translated text.
    """
    # Load the tokenizer and model
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True)

    # Generate the translation
    translated = model.generate(**inputs)

    # Decode and return the translated text
    output_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return output_text
```

# Language model dictionary for English to target language

```python
models = {
    "Spanish": "Helsinki-NLP/opus-mt-en-es",
    "French": "Helsinki-NLP/opus-mt-en-fr",
    "German": "Helsinki-NLP/opus-mt-en-de",
    "Chinese": "Helsinki-NLP/opus-mt-en-zh",
    "Arabic": "Helsinki-NLP/opus-mt-en-ar",
    "Hindi": "Helsinki-NLP/opus-mt-en-hi",
    "Japanese": "Helsinki-NLP/opus-mt-en-ja",
    "Russian": "Helsinki-NLP/opus-mt-en-ru",
    "Italian": "Helsinki-NLP/opus-mt-en-it",
}
```

# Display supported languages

# Get user input for language and text

 # Input text to translate

 # Translate and display the result

```python
print("Supported Languages:")
for lang in models.keys():
    print(f"- {lang}")

target_language = input("\nEnter the target language (e.g., Spanish, French): ").strip()
model_name = models.get(target_language, None)

if model_name:
    input_text = input("Enter text in English: ").strip()
    translated_text = translate_text(input_text, model_name)
    print(f"\nTranslated Text in {target_language}: {translated_text}")
else:
    print("Sorry, the language is not supported.")

```
