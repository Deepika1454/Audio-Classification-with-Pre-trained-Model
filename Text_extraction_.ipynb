```python
import easyocr
import matplotlib.pyplot as plt
from PIL import Image
reader = easyocr.Reader(['en'])  # 'en' = English
image_path = "/content/canva-blue-grey-minimalist-modern-business-card-fNjrFt342kM.jpg"
result = reader.readtext(image_path)
img = Image.open(image_path)
plt.imshow(img)
plt.axis('off')
plt.title('Detected Text')
plt.show()
for (bbox, text, prob) in result:
    print(f"üìù {text} (Confidence: {prob:.2f})")
```
