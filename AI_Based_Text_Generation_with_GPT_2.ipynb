 Load the Pretrained GPT-2 Model

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")
```

Generate Text

```python
prompt = "A princess named with sarah lived in a palace,"
generated_text = generator(prompt, max_length=100, num_return_sequences=1)
print(generated_text[0]["generated_text"])
```

Change max_length to make text longer/shorter.
Use a higher temperature for more creative text

```python
generator(prompt, max_length=100, num_return_sequences=1, temperature=0.9)
```
