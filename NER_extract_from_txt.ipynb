```python

!pip install spacy pandas
!python -m spacy download en_core_web_sm

import spacy
import pandas as pd
from google.colab import files

nlp = spacy.load("en_core_web_sm")

print("Please upload a .txt file:")
uploaded = files.upload()

filename = list(uploaded.keys())[0]
with open(filename, "r", encoding="utf-8") as file:
    text = file.read()


def extract_ner_to_csv(text):
    if text:
        doc = nlp(text)
        entities = [(ent.text, ent.label_) for ent in doc.ents]


        df = pd.DataFrame(entities, columns=["Entity", "Label"])


        df.to_csv("NER_Extracted.csv", index=False)
        print("\nNamed Entities Extracted and Saved as 'NER_Extracted.csv'\n")
        print(df.head())

if text:
    print("\nUploaded File Content Preview (First 500 chars):\n", text[:500])
    extract_ner_to_csv(text)
else:
    print("No valid text found in the uploaded file.")

files.download("NER_Extracted.csv")
```
